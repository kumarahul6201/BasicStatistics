```python
# Step 1: Load the Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('sales_data_with_discounts.csv')

# Display the first few rows of the dataset
display(data.head())

```

```python
# Step 2: Identify Numerical Columns
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns.tolist()
print("Numerical Columns:", numerical_columns)
```

```python
# Step 3: Calculate Statistics (Mean, Median, Mode, Standard Deviation)
statistics = pd.DataFrame(index=numerical_columns)

statistics['Mean'] = data[numerical_columns].mean()
statistics['Median'] = data[numerical_columns].median()
statistics['Mode'] = data[numerical_columns].mode().iloc[0]
statistics['Standard Deviation'] = data[numerical_columns].std()

# Display the statistics
display(statistics)
```

```python
# Step 4: Interpretation of Statistics
for column in numerical_columns:
    print(f"Column: {column}")
    print(f"Mean: {statistics.loc[column, 'Mean']}")
    print(f"Median: {statistics.loc[column, 'Median']}")
    print(f"Mode: {statistics.loc[column, 'Mode']}")
    print(f"Standard Deviation: {statistics.loc[column, 'Standard Deviation']}")
    print("\n")
```

```python
# Step 1: Load the Dataset and Identify Numerical Columns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('sales_data_with_discounts.csv')

# Display the first few rows of the dataset
display(data.head())

```

```python
# Identify numerical columns
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns.tolist()
print("Numerical Columns:", numerical_columns)
```

```python
# Step 2: Standardize the Numerical Columns
def standardize_column(column):
    return (column - column.mean()) / column.std()

# Standardize the numerical columns and create a new DataFrame
data_standardized = data.copy()
for column in numerical_columns:
    data_standardized[column] = standardize_column(data[column])

# Display the first few rows of the standardized dataset
display(data_standardized.head())
```

```python
# Step 3: Show Before and After Comparisons of the Data Distributions
def plot_distributions(data, data_standardized, columns):
    for column in columns:
        plt.figure(figsize=(14, 6))
        
        # Original data distribution
        plt.subplot(1, 2, 1)
        sns.histplot(data[column], kde=True)
        plt.title(f'Original Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        
        # Standardized data distribution
        plt.subplot(1, 2, 2)
        sns.histplot(data_standardized[column], kde=True)
        plt.title(f'Standardized Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        
        plt.tight_layout()
        plt.show()
```

```python
# Plot before and after comparisons for numerical columns
plot_distributions(data, data_standardized, numerical_columns)
```

```python
# Step 1: Load the Dataset and Identify Categorical Columns
import pandas as pd

# Load the dataset
data = pd.read_csv('sales_data_with_discounts.csv')

# Display the first few rows of the dataset
display(data.head())

# Identify categorical columns
categorical_columns = data.select_dtypes(include=['object', 'category']).columns.tolist()
print("Categorical Columns:", categorical_columns)

```
